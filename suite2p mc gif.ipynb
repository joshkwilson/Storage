```python
import imageio
import os

# Set the directory path to your tif files

folder_name = r'D:\JW\03\033023\suite2p\plane0'

im_name = 'reg_tif'

dir_path = os.path.join(folder_name, im_name) 
```

```python
# Get the list of tif files in the directory
tif_files = [f for f in os.listdir(dir_path) if f.endswith('.tif')]

# Create a list of image arrays from the tif files
images = []
for tif_file in tif_files:
    image = imageio.imread(os.path.join(dir_path, tif_file))
    images.append(image)

# Create the gif file that will save in cwd
imageio.mimsave('animation.gif', images, fps=10)
```
